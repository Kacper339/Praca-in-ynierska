# Wyniki uzyskanych sieci

```python
import os
import numpy as np
import matplotlib.pyplot as plt
from skimage.transform import resize
from tensorflow.keras.models import load_model
from tensorflow.keras.utils import to_categorical
from sklearn.metrics import accuracy_score, confusion_matrix
import seaborn as sns
from sklearn.metrics import precision_score, recall_score, f1_score
```

```python
test_dir = "seg_test/seg_test"
x_test = []
y_test = []

folderlist = os.listdir(test_dir)
folderlist.sort()
class_names = folderlist

for i, category in enumerate(folderlist):
    pictures = os.listdir(test_dir + "/" + category)
    for picture in pictures:
        img = plt.imread(test_dir + "/" + category + '/{0}'.format(picture), 0)
        img_resized = resize(img, (228, 228))
        x_test.append(img_resized)
        y_test.append(i)
```

```python
x_test = np.array(x_test, 'float32')
y_test = np.array(y_test, 'float32')
```

```python
x_test_norm = (x_test - np.min(x_test)) / (np.max(x_test) - np.min(x_test)) * 255.0
x_test_norm = x_test_norm / 255.0
```

```python
y_test_oneHot = to_categorical(y_test, num_classes=6)
```

```python
def evaluate_model(model, x_test_norm, y_test_oneHot, model_name):
    loss, accuracy = model.evaluate(x_test_norm, y_test_oneHot, verbose=0)
    print(f"{model_name} - Loss: {loss:.4f}, Accuracy: {accuracy:.4f}")
    
    y_pred = np.argmax(model.predict(x_test_norm), axis=1)
    y_true = np.argmax(y_test_oneHot, axis=1)
    
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=class_names, yticklabels=class_names)
    plt.title(f'Macierz pomyłek - {model_name}')
    plt.ylabel('Prawdziwe etykiety')
    plt.xlabel('Przewidywane etykiety')
    plt.savefig(f'macierz_pomylek_{model_name}.png')
    plt.show()
```

```python
from sklearn.manifold import TSNE
from tensorflow.keras.models import Model
```

```python
def tSNE(model, x_data, y_data, title):
    feature_model = Model(inputs=model.input, outputs=model.layers[-2].output)
    features_test = feature_model.predict(x_data)
    
    tsne = TSNE(n_components=2, random_state=0)
    tsne_features = tsne.fit_transform(features_test)
    
    plt.figure(figsize=(10, 8))
    sns.scatterplot(x=tsne_features[:, 0], y=tsne_features[:, 1], hue=[class_names[int(i)] for i in y_data],
                    palette='tab10', s=60, alpha=0.8, edgecolor="k", legend="full")
    
    plt.title(f'Wizualizacja t-SNE: {title}')
    plt.legend(title='Klasy', loc='best', bbox_to_anchor=(1.05, 1), borderaxespad=0)
    plt.savefig(f'Wizualizacja_tSNE_{title}.png', bbox_inches='tight')
    plt.show()
```
